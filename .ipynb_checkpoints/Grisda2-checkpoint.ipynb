```python
#URLs
main_url= 'https://www.grisda.org'
url1= 'https://www.grisda.org/articles?primary=geology&secondary=geologic-processes'
url2= 'https://www.grisda.org/articles?page=2&secondary=geologic-processes'
url3= 'https://www.grisda.org/articles?page=3&secondary=geologic-processes'
url4= 'https://www.grisda.org/articles?page=4&secondary=geologic-processes'

#Get info from grid boxes
def getInfo(url):
    r= requests.get(url)
    soup= BeautifulSoup(r.text, 'html.parser')
    #Get grid with title and link
    grids= soup.section.find('div', {'id': 'article'}).div.div.find_all('div', class_='grid-line')
    
    #Get title, author, date and link
    info= []
    for grid in grids:
        info.append({
            'title_author': grid.h3.text,            
            'date': grid.time.text,
            'link': grid.a['href']
        })
    
    return info

#Get title and article 
def getArticle(page_n):
    r= requests.get(main_url+page_n['link'])
    soup= BeautifulSoup(r.text, 'html.parser')
    page= soup.find('div', class_='page')
    
    #Return None
    if page == None:
        return None
    
    #Remove figures
    for fig in page.select('figure'):
        fig.extract()
        
    #Get article body
    tex= ''
    for info in page.find_all('p'):
        tex+= info.text
    
    #Split by footer
    tex= re.split('ENDNOTES|Endnotes|REFERENCES|References|Footnotes|FOR FURTHER STUDY|LITERATURE CITED', tex)[0]
    
    #Remove first cite
    bib_pat= r'(.+doi.+(Summary\. ?|\n)|.+DOI.+(Summary\. ?|\n))'
    tex= re.sub(bib_pat,'', tex)
        
    #Remove cites
    cites_pat= r'\[[\w\.;,\- ]+\]'
    tex= re.sub(cites_pat,'', tex)
    
    #Remove parenthesis
    pars_pat= r'\([\w\.;,\- ]+\)'
    tex= re.sub(pars_pat,'', tex)
    
    #Remove Figure mentions
    figs_pat= r'[F|f]ig[\.\w]+ [\w,\- ]+|see [F|f]ig[\.\w]+ [\w,\- ]+'
    tex= re.sub(figs_pat,'', tex)
    
    #Remove stop words
    strwords= r'WHAT THIS ARTICLE IS ABOUT|ABSTRACT|INTRODUCTION|ACKNOWLEDGMENTS?|CONCLUSIONS?|Conclusions?|SUMMARY|DISCUSSION|Geoscience Research Institute'
    tex= re.sub(strwords,'', tex)
            
    #Get title and article body
    doc= {
        'title': page_n['title_author'].split('by ')[0][:-1],
        'author': page_n['title_author'].split('by ')[-1],
        'doc': tex
         }
    
    return doc

def normalize(tex):
    #Lowercase
    tex= tex.lower()
    #Remove punctuation
    tex= re.sub(r'[^\w\s]','', tex)
    #Remove white spaces
    texf= tex.strip()
    #Tokenize
    #tokens= word_tokenize(tex)
    #textf= nltk.Text(tokens)
    
    return texf
```

```python
#Get grid boxes
page1= getInfo(url1)
page2= getInfo(url2)
page3= getInfo(url3)
page4= getInfo(url4)

#Get articles
docs= []
for pagen in [page1]:#, page2, page3, page4]:
    for page in pagen:
        docs.append(getArticle(page))

#Remove None
docs= list(filter(None, docs))
#Remove biblio
docs[8]['doc']= re.split('Ariel A. Roth', docs[8]['doc'])[0]
#CHECK BY ICELAND
#Remove film review
del docs[4]
```

```python
"""
2.1 Similarity Flood ADJs
Higher: global-worldwide: .68, local-traditional: .44, global-cathastrophic: .41
"""

flood_adjs= FreqDist(all_nouns).most_common(6)
only_adjs= [x[0].split()[0] for x in flood_adjs]
print(only_adjs)
print()

all_sims= []
for adx in only_adjs:
    doc1= nlp(adx)
    sim= []
    
    for ady in only_adjs:
        doc2= nlp(ady)
        sim.append(doc1.similarity(doc2))
        
    all_sims.append(sim)

arr= np.array(all_sims)
abr= [x[0:4] for x in only_adjs]
df= pd.DataFrame(data=arr, index=abr, columns=abr)
print(df)
```

```python
"""
2.3 Phrase Matcher - Root
VERB: Earth expanding 25
"""

nlp= spacy.load("en_core_web_md")

matcher= Matcher(nlp.vocab)
pattern= [{"POS": "VERB"}]
matcher.add("PLACE", [pattern])

alls= []
for doc1 in docs:
    doc= nlp(doc1['doc'])
    doc.ents= []
    
    for match_id, start, end in matcher(doc):
        span= Span(doc, start, end, label="VERB")

        doc.ents = list(doc.ents) + [span]
        span_root_head= span.root.head
        if(span_root_head.text != span.text and len(span.text) > 4 and len(span_root_head) > 4):
            alls.append(span_root_head.text+" --> "+span.text)
            #print(span_root_head.text, "-->", span.text)

FreqDist(alls).most_common(15)
```

```python
"""
3.1 Pipe: ELEMENTS
Found: mobile ELEMENTS
"""

nlp= spacy.load("en_core_web_sm")

elems= ['mafic', 'felsic', 'intermediate', 'mobile', 'immobile', 'alkaline']
elem_patterns= list(nlp.pipe(elems))
print("elements_patterns:", elem_patterns)

name= 'ELEMENTS'
matcher= PhraseMatcher(nlp.vocab)
matcher.add(name, elem_patterns)

@Language.component('el_component')
def el_component(doc):
    matches= matcher(doc)
    spans= [Span(doc, start, end, label=name) for match_id, start, end in matches]
    doc.ents= spans
    
    return doc

nlp.add_pipe('el_component', after="ner")
print()

li= []
for doc in docs:
    doco= nlp(doc['doc'])
    li.append([(ent.text, ent.label_) for ent in doco.ents])

#Flat list
li= [item for sublist in li for item in sublist]
print(li)
```

```python
"""
4.2 DISASTER ADJs
Found: Lisbon earthquake, Alaska earthquake, Chilean earthquake
"""

elems= ['earthquake', 'earthquakes', 'tsunami', 'tsunamis', 'volcanic eruption']

for doc in docs:
    doco= nlp(doc['doc'])
    
    for ent in doco:
        if (ent.ent_type_ == 'DISASTER'):
            if (ent.text in elems):
                if(len(doco[ent.i-1].text) > 5):
                    print(doco[ent.i-1].text, ent.text)
```

```python
"""
4.3 DISASTER - VERB
Found: earthquake shook, tsunami triggered
"""

alls= []
ents2= []
for doc1 in docs:
    
    doc= nlp(doc1['doc'])
    
    matcher= Matcher(nlp.vocab)
    patterns= [{"ENT_TYPE": "DISASTER"}]
    matcher.add("DIS2", [patterns])
    matches= matcher(doc)
    
    for match_id, start, end in matches:
        span= Span(doc, start, end, label="DIS2")
        ents2.append(span.text)
        span_root_head= span.root.head
        #print('Root', span_root_head.pos_)
        if (span.text in elems):
            if(span_root_head.text != span.text and len(span.text) > 4 and len(span_root_head) > 4):
                if(span_root_head.pos_ == 'VERB'):
                    alls.append(span.text+" --> "+span_root_head.text)
                    print(span_root_head.text, "-->", span.text)

FreqDist(alls).most_common(15)
```

```python
"""
4.4 DISASTER and GPE/LOC
GPE Found: Deccan volcanism and the Chicxulub impact, earthquake shook Mexico City
LOC Found: earthquake shook the New England coast, eruption of Mount St. Helens, River Valley was raised by volcanic debris,
Old Testament mentions an earthquake that occurred in the days of King Uzziah
"""

matcher= Matcher(nlp.vocab)
pattern= [{"ENT_TYPE": "LOC"}, {"OP": "?"}, {"OP": "?"}, {"OP": "?"}, {"OP": "?"}, {"ENT_TYPE": "DISASTER"}, {"OP": "?"}, {"OP": "?"}, {"OP": "?"}]

matcher.add("DISPLAC", [pattern])

all_placs= []
for doc in docs:
    mat= nlp(doc['doc'])
    matches= matcher(mat)
    
    if matches:
        print(doc['title'])
        for match_id, start, end in matches:
            all_placs.append(mat[start:end].text)
            
        print([mat[start:end].text for match_id, start, end in matches])
        print()

elems2= ['earthquake', 'earthquakes', 'tsunami', 'tsunamis', 'volcanic', 'eruption', 'volcanism']

fina= []
for plac in all_placs:
    for d in elems2:
        if d in plac:
            fina.append(plac)
            
FreqDist(fina).most_common()
```

```python
from pyspark.sql.functions import monotonically_increasing_id

titles = titles.withColumn("id", monotonically_increasing_id())
predictions = predictions.withColumn("id", monotonically_increasing_id())

df3 = titles.join(predictions, "id", "outer").drop("id")
```

```python
#Get by cluster
c0= df3.filter(df3.prediction == 0).select(df3.title)
c1= df3.filter(df3.prediction == 1).select(df3.title)
c2= df3.filter(df3.prediction == 2).select(df3.title)
```

```python
def getLabels(c):
    li0= [row for row in c.select("title").collect()]
    alls0= []
    for i in li0:
        al= []
        for j in i:
            al.append(j)
        alls0.append(al)
    #Flat list
    alls0= [item for sublist in alls0 for item in sublist]
    #-by
    alls0= [a.split('by')[0][:-1] for a in alls0]
    
    return alls0

print(getLabels(c0))
print(getLabels(c1))
print(getLabels(c2))
```

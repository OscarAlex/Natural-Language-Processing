```python
from bs4 import BeautifulSoup
import re
import requests

import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

from itertools import groupby
from collections import Counter

from gensim.summarization.summarizer import summarize
from gensim.summarization.textcleaner import split_sentences
import heapq
from nltk.sem import extract_rels, rtuple
```

```python
#URLs
main_url= 'https://www.grisda.org'
url1= 'https://www.grisda.org/articles?primary=geology&secondary=geologic-processes'
url2= 'https://www.grisda.org/articles?page=2&secondary=geologic-processes'
url3= 'https://www.grisda.org/articles?page=3&secondary=geologic-processes'
url4= 'https://www.grisda.org/articles?page=4&secondary=geologic-processes'

#Get info from grid boxes
def getInfo(url):
    r= requests.get(url)
    soup= BeautifulSoup(r.text, 'lxml')
    #Get grid with title and link
    grids= soup.section.find('div', {'id': 'article'}).div.div.find_all('div', class_='grid-line')
    
    #Get title, author, date and link
    info= []
    for grid in grids:
        info.append({
            'title_author': grid.h3.text,            
            'date': grid.time.text,
            'link': grid.a['href']
        })
    
    return info

#Get title and article 
def getArticle(page_n):
    r= requests.get(main_url+page_n['link'])
    soup= BeautifulSoup(r.text, 'lxml')
    page= soup.find('div', class_='page')
    
    #Return None
    if page == None:
        return None
    
    #Remove figures
    for fig in page.select('figure'):
        fig.extract()
        
    #Get article body
    tex= ''
    for info in page.find_all('p'):
        tex+= info.text
    
    #Split by footer
    tex= re.split('ENDNOTES|Endnotes|REFERENCES|References|Footnotes|FOR FURTHER STUDY|LITERATURE CITED', tex)[0]
    
    #Remove first cite
    bib_pat= r'(.+doi.+(Summary\. ?|\n)|.+DOI.+(Summary\. ?|\n))'
    tex= re.sub(bib_pat,'', tex)
        
    #Remove cites
    cites_pat= r'\[[\w\.;,\- ]+\]'
    tex= re.sub(cites_pat,'', tex)
    
    #Remove parenthesis
    pars_pat= r'\([\w\.;,\- ]+\)'
    tex= re.sub(pars_pat,'', tex)
    
    #Remove Figure mentions
    figs_pat= r'[F|f]ig[\.\w]+ [\w,\- ]+|see [F|f]ig[\.\w]+ [\w,\- ]+'
    tex= re.sub(figs_pat,'', tex)
    
    #Remove stop words
    strwords= r'WHAT THIS ARTICLE IS ABOUT|ABSTRACT|INTRODUCTION|ACKNOWLEDGMENTS?|CONCLUSIONS?|Conclusions?|SUMMARY|DISCUSSION|Geoscience Research Institute'
    tex= re.sub(strwords,'', tex)
            
    #Get title and article body
    doc= {
        'title': page_n['title_author'],
        'doc': tex
         }
    
    return doc

def normalize(tex):
    #Lowercase
    tex= tex.lower()
    #Remove punctuation
    tex= re.sub(r'[^\w\s\.]','', tex)
    #Remove white spaces
    tex= tex.strip()
    #Tokenize
    tokens= word_tokenize(tex)
    textf= nltk.Text(tokens)
    
    return textf
```

```python
#Get grid boxes
page1= getInfo(url1)
page2= getInfo(url2)
page3= getInfo(url3)
page4= getInfo(url4)

#Get articles
docs= []
for pagen in [page1, page2, page3, page4]:
    for page in pagen:
        docs.append(getArticle(page))

#Remove None
docs= list(filter(None, docs))
#Remove biblio
docs[8]['doc']= re.split('Ariel A. Roth', docs[8]['doc'])[0]
#Remove film review
del docs[4]
```

```python
#Tokens per article
tokens= []
for doc in docs:
    tokens.append(normalize(doc['doc']))

#Lemmatizer
wnl= nltk.WordNetLemmatizer()
#Lemms per article
tokens_lems= []
#All lemms
all_lems= []
#Get lemms
for tt in tokens:
    lems= []
    for t in tt:
        lems.append(wnl.lemmatize(t))
        all_lems.append(wnl.lemmatize(t))
    tokens_lems.append(nltk.Text(lems))
```

```python
"""
5.1 Tags per article
Nouns are the most frequent in articles, 
then verbs or adjective.
"""

def tags_article(token):
    #Get tags
    all_tags= nltk.pos_tag(token, tagset='universal')
    #Sort by tag
    sorted_tags= sorted(all_tags, key=lambda x: x[1])

    #Group by tags
    list_tags= []
    for key, group in groupby(sorted_tags, lambda x: x[1]):
        list_tags.append(list(group))
    #Flat list
    list_tags= [item for sublist in list_tags for item in sublist]
    
    return print(Counter(elem[1] for elem in list_tags))

for doc, token in zip(docs, tokens):
    print(doc['title'])
    tags_article(token)
    print()
```

```python
"""
5.2 Word tags: dinosaur
Dinosaur, flood is used as a noun or adjective.
"""

def different_tags(token, word):
    all_tags= nltk.pos_tag(token, tagset='universal')
    cfd1= nltk.ConditionalFreqDist(all_tags)
    
    #cfd1[word].most_common()
    if len(cfd1[word].most_common()) >= 1:
        print(cfd1[word].most_common())
    
for doc, token in zip(docs, tokens):
    #print(doc['title']) if different_tags(token, 'dinosaur') else print()
        #different_tags(token, 'dinosaur')
    print(doc['title'])
    different_tags(token, 'dinosaur')
    print()
```

```python
"""
5.3 Specific tag: Adjectives
Found: global, large, great, physical, extraterrestrial.
"""

def specific_tag(token, tagg):
    wsj= nltk.pos_tag(token, tagset='universal')
    word_tag_fd= nltk.FreqDist(wsj)
    return [word for (word, tag) in word_tag_fd if tag == tagg]

for doc, token in zip(docs[:40], tokens[:40]):
    print(doc['title'])
    print(specific_tag(token, 'ADJ'))
    print()
```

```python
"""
6.1 Summarizer: Scores
Assign a score to each token and sentence based on frequency.
Get the sentences with the highest score.
"""

def summary2(doc, no):
    cfd1= nltk.FreqDist(word_tokenize(doc))
    max_common= cfd1.most_common(1)[0][1]
    #Words score
    new= {}
    #Kind of normalization
    for word in cfd1.keys():
        new[word]= (cfd1[word]/max_common)

    #Sentences score
    sentences_score= {}
    #Add score to sentences
    for sentence in split_sentences(doc):
        for word in word_tokenize(sentence):
            if word in new.keys():
                if (len(sentence.split(" "))) < 30:
                    if sentence not in sentences_score.keys():
                        sentences_score[sentence]= new[word]
                    else:
                        sentences_score[sentence]+= new[word]

    #Return the sentences with the highest score
    summary= heapq.nlargest(no, sentences_score, key=sentences_score.get)
    print(" ".join(summary))

for doc in docs:
    print(doc['title'])
    print(summary2(doc['doc'], 10))
    print()
```

```python
"""
6.2 Summarizer: Gensim
- Pre-process the given text.
- Make a graph with sentences are the vertices.
- The graph has edges denoting the similarity between the two sentences at the vertices.
- Run PageRank algorithm on this weighted graph.
- Pick the highest scoring vertices and append them to the summary.
"""

def summary(doc, percentage):
    #If more than 10 sentences
    if len(split_sentences(doc)) > 10:
        test_summary= summarize(doc, ratio= percentage, split=False)
    else:
        test_summary= doc
        
    print(test_summary)

for doc in docs:
    print(doc['title'])
    summary(doc['doc'], 0.1)
    print()
```

```python
"""
7.1 Chunks: NP
Found: solar radiation, solar wind
"""

def nps(doc):
    #Sentences per article
    sentences= []
    for sentence in doc.split('.'):
        sentences.append(word_tokenize(sentence))

    #Tags per sentence per article
    tags= []
    for sentence in sentences:
        tags.append(nltk.pos_tag(sentence, tagset='universal'))

    #Remove empty lists
    tags= [item for item in tags if item]
    
    grammar= "NP: {<DET>?<ADJ>*<NOUN>+}"
    cp= nltk.RegexpParser(grammar)
    
    def filt(x):
        return x.label() == 'NP'

    for tag in tags:
        result= cp.parse(tag)
        #Subtrees
        for subtree in result.subtrees(filter= filt): 
            if (len(subtree) > 1):
                if (len(subtree) == 2 and subtree[0][1] != 'DET'):
                    print(subtree)
                elif(len(subtree) > 2):
                    print(subtree)
        
for doc in docs:
    print(doc['title'])
    nps(doc['doc'])
    print()
```

```python
"""
7.2 Chunks: QUAN
Found: Expanding Earth related to day's duration
"""

#Tags per article
def nps(doc):
    #Sentences per article
    sentences= []
    for sentence in doc.split('.'):
        sentences.append(word_tokenize(sentence))

    #Tags per sentence per article
    tags= []
    for sentence in sentences:
        tags.append(nltk.pos_tag(sentence, tagset='universal'))
    
    #Remove empty lists
    tags= [item for item in tags if item]
    
    grammar= "QUAN: {<....?><....?><....?><....?><....?><NUM><NOUN><....?><....?><....?><....?><....?><....?>}"
    cp= nltk.RegexpParser(grammar)
    
    def filt(x):
        return x.label() == 'QUAN'

    for tag in tags:
        result= cp.parse(tag)
        #Subtrees
        for subtree in result.subtrees(filter= filt):
            print(subtree[0][0],subtree[1][0],subtree[2][0],subtree[3][0],subtree[4][0],subtree[5][0],
                  subtree[6][0],subtree[7][0],subtree[8][0],subtree[9][0])
            
        
for doc in docs:
    print(doc['title'])
    nps(doc['doc'])
    print()
```

```python
"""
7.3 Entities: Facility
Found: Grand Canyon, Staircase
"""
#Check
def filt_g(x):
    #GPE PERSON LOCATION ORGANIZATION FACILITY
    return x.label() == 'FACILITY'

ents= []
for doc in docs:
    entities= nltk.ne_chunk(nltk.pos_tag(word_tokenize(doc['doc'])))
    
    for subtree in entities.subtrees(filter= filt_g):
        if (len(subtree) > 1):
            ents.append(' '.join(c[0] for c in subtree))

FreqDist(ents).most_common(20)
```

```python
"""
7.4 Entities: Person
Found: Lake Tanganyika, Missoula, Gosiute
Spirit Lake
Mount St. Helens
"""

def filt_g(x):
    #GPE PERSON LOCATION ORGANIZATION FACILITY
    return x.label() == 'PERSON'

subs= []
for doc in docs:
    entities= nltk.ne_chunk(nltk.pos_tag(word_tokenize(doc['doc'])))
    
    for subtree in entities.subtrees(filter= filt_g):
        if (len(subtree) > 1):
            subs.append(subtree)
            #print(subtree)
    
li= []
for sub in subs:
    for su in sub:
        li.append(su)

FreqDist([w for w in li if (len(w[0]) > 3)]).most_common(20)
```

```python
"""
7.5 Relation:
IN: 'God created objects in the Creation Week.' Person Organization
    'Volcanism in Iceland' Person GPE
"""
OF= re.compile(r'.*\bof\b.*')
IN= re.compile(r'.*\bin\b.*')

for doc in docs:
    sentences= nltk.sent_tokenize(doc['doc'])
    tokenized_sentences= [nltk.word_tokenize(sentence) for sentence in sentences]
    tagged_sentences= [nltk.pos_tag(sentence) for sentence in tokenized_sentences]
    
    #print(doc['title'])
    for sent in tagged_sentences:
        sent= nltk.ne_chunk(sent)
        #GPE PERSON LOCATION ORGANIZATION FACILITY
        rels= extract_rels('PERSON', 'GPE', sent, corpus='ace', pattern= IN)
        
        for rel in rels:
            print(rtuple(rel))
    #print()
```

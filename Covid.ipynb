```python
import json
```

```python
def get_unique(json_file):
    with open(json_file) as input_file:
        data= json.load(input_file)

    for i in range(len(data['cov'])):
        sentence= data['cov'][i]['doc']
        st= ' '.join(sentence)
        data['cov'][i]['doc']= st

    res= list(map(dict, set(tuple(sorted(sub.items())) for sub in data['cov'])))
    
    return res
```

```python
c1= get_unique('covid1.json')
c2= get_unique('covid2.json')
c3= get_unique('covid3.json')
c4= get_unique('covid4.json')
c5= get_unique('covid5.json')
```

```python
all_di= [c1,c2,c3,c4,c5]
flat_list = [item for sublist in all_di for item in sublist]

covid400= {'covid400': flat_list}
#print(all_di[0])
print(len(c1))
print(len(c2))
print(len(c3))
print(len(c4))
print(len(c5))
print(len(flat_list))
```

```python
#Save json
with open('covid400.json', 'w') as fp:
    json.dump(covid400, fp)
```

```python
#Open json
with open('covid400.json') as input_file:
    data = json.load(input_file)
    
print(data['covid400'][0])
```

```python
import nltk
from nltk.tokenize import word_tokenize

def normalize(tex):
    #Lowercase
    #tex= tex.lower()
    #Remove punctuation
    #tex= re.sub(r'[^\w\s\.]','', tex)
    #Remove white spaces
    tex= tex.strip()
    #Tokenize
    tokens= word_tokenize(tex)
    textf= nltk.Text(tokens)
    
    return textf
```

```python
#VACCINE
#Tokens per article
tokens= []
for doc in data['covid400']:
    tokens.append(normalize(doc['doc']))
```

```python
"""
Collocations: 
Influenza vaccine, vaccine efficacy, sars vaccine
"""

for tok in tokens:
    tok.collocations()
    print()
    #print(token.collocations())
```

```python
"""
Concordance: 

"""

for tok in tokens:
    tok.concordance('vaccine')
    print()
```

```python
"""
Tokenized text: lack of

"""

for tok in tokens:
    tok.findall(r"<\w+>{20} <vaccine> <\w+>{20}")
    print()
```
